#### Subqueries with Pandas

```python
#1.

import pandas as pd

# Perform INNER JOIN based on 'Movie Name' column
merged_df = pd.merge(df1, df2, on='Movie Name', how='inner')

# Display all details from both tables where movie names are the same
print(merged_df)
```

```python
#2.

merged_df= pd.merge(df1, df2, on='Movie Name', how='inner')

filter_df=merged_df[merged_df['Genre'].str.contains('Action', case=False)]

print(filter_df[['Movie Name','Director1','Director2','Rating','Gross Collection']])
```

```python
#3.


import pandas as pd

# Perform INNER JOIN based on 'Movie Name' column
merged_df = pd.merge(df1, df2, on='Movie Name', how='inner')

# Calculate the maximum gross collection value
max_gross_collection = merged_df['Gross Collection'].max()

# Filter the DataFrame to get the rows with the highest gross collection
filter_df = merged_df[merged_df['Gross Collection'] == max_gross_collection]

print(filter_df)
```

```python
#4.

import pandas as pd

# Perform INNER JOIN based on 'Movie Name' column
merged_df = pd.merge(df1, df2, on='Movie Name', how='inner')

# Calculate the maximum rating
max_rating = merged_df['Rating'].max()

# Filter the DataFrame to get the rows with the highest gross collection
filter_df = merged_df[merged_df['Rating'] == max_rating]

print(filter_df)
```

```python
#5.

import pandas as pd

# Perform INNER JOIN based on 'Movie Name' column
merged_df = pd.merge(df1, df2, on='Movie Name', how='inner')

# Find the movie with the lowest gross collection in the merged DataFrame
lowest_gross = merged_df['Gross Collection'].min()

# Find the movie with the lowest rating in the merged DataFrame
lowest_rating = merged_df['Rating'].min()

# Filter the DataFrame to get the rows with the lowest gross collection and lowest ratings
result_df = merged_df[(merged_df['Gross Collection'] == lowest_gross) & (merged_df['Rating'] == lowest_rating)]

# Display all details from both tables with the lowest gross collection and lowest ratings
print(result_df)
```
